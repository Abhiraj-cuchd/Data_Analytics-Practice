# Introduction to MatPlotLib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 4
```

```python
x
```

```python
y
```

```python
#Functional Method
plt.plot(x,y, 'r--')
plt.xlabel('Population')
plt.ylabel('Malnutrtion')
plt.title('BMI Index')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.xlabel('Population')
plt.ylabel('Malnutrition')
plt.title('BMI Index-INDIA')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
plt.xlabel('Population')
plt.ylabel('Malnutrition')
plt.title('BMI Index-PAKISTAN')
```

```python
#Object Oriented API Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y, 'y*--')
axes.set_xlabel('Population')
axes.set_ylabel('Malnutrition')
axes.set_title('BMI INDEX - INDIA')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r--')
axes1.set_title('LARGER PLOT')
axes2.plot(y,x,'g*-')
axes2.set_title('SMALLER PLOT')
```

```python
fig , axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y,'r*-')
axes[0].set_title('India')

axes[1].plot(y,x,'g*--')
axes[1].set_title('Pakistan')

plt.tight_layout()
```

# Figure Size, Aspect Ratio and DPI

```python
fig, axes = plt.subplots(nrows = 2, ncols=1, figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_pic.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x, x**2,label = 'x-square')
ax.plot(x, x**3, label = 'x Cubed')

ax.legend(loc = 0)
```

Plot Appeareance

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,y, color='purple', lw=2, ls = '--', marker = 'o', markersize = 10)
```
